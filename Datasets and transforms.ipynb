#### Create a Dataset Class <br/>

using the components in the last section to build a dataset class and then create an object.

```python
# Create your own dataset object

class Dataset(Dataset):

    # Constructor
    def __init__(self, csv_file, data_dir, transform=None):

        # Image directory (this will be the directory containing the CSV and potentially images)
        self.data_dir=data_dir
        # The transform is going to be used on image
        self.transform = transform
        # Construct the full path to the CSV file
        # The CSV file is expected to be directly in data_dir, not in a subfolder like 'img'
        csv_file_path = os.path.join(self.data_dir, csv_file)

        # Load the CSV file contains image info
        # error handling in case the file is not found
        if not os.path.exists(csv_file_path):
            raise FileNotFoundError(f"CSV file not found: {csv_file_path}")

        self.data_name = pd.read_csv(csv_file_path, header=0)
        # Number of images in dataset
        self.len=self.data_name.shape[0]

    # Get the length
    def __len__(self):
        return self.len

    # Getter
    def __getitem__(self, idx):

        # The class label for the image (first column, index 0)
        y = self.data_name.iloc[idx, 0]
        # Get the pixel data for the image (all columns from index 1 onwards)
        # Convert to numpy array and ensure data type is uint8 (common for image data)
        image_data = self.data_name.iloc[idx, 1:].values.astype('uint8')
        # Reshape the 1D array of pixel data into a 28x28 image array
        image_array = image_data.reshape((28, 28))

        # If there is any transform method, apply it onto the image
        # The transform should be applied to the image_array (numpy array) or a PIL Image
        image = Image.fromarray(image_array, 'L') # 'L' for grayscale

        if self.transform:
            image = self.transform(image)

        return image, y
```

```python
# Create the dataset objects
# Define the CSV file name
csv_file = "fashion-mnist_train.csv" # Or "fashion-mnist_test.csv" depending on which dataset

# Use the extracted directory where the CSVs are located
# The extracted directory is /content/fashionmnist based on previous steps
dataset = Dataset(csv_file=csv_file, data_dir=extracted_dir)

print("Dataset object created successfully.")
```

```python
import torchvision.transforms as transforms
```

apply some image transform functions on the dataset object. The iamge can be cropped and converted to a tensor. I will use <code>transform.Compose</code> I learned from previous sections to combine the two transform functions.


```python
# Combine two transforms: crop and convert to tensor. Apply the compose to MNIST dataset

croptensor_data_transform = transforms.Compose([transforms.CenterCrop(20), transforms.ToTensor()])
dataset = Dataset(csv_file=csv_file, data_dir=extracted_dir, transform=croptensor_data_transform )
print("The shape of the first element tensor: ", dataset[0][0].shape)
```

```python
# Construct the compose. Apply it on MNIST dataset. Plot the image out.

fliptensor_data_transform = transforms.Compose([transforms.RandomVerticalFlip(p=1),transforms.ToTensor()])
dataset = Dataset(csv_file=csv_file , data_dir=extracted_dir,transform=fliptensor_data_transform )
show_data(dataset[0])
```
